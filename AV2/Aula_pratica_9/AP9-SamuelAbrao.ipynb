```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('diamonds.csv')
```

### Exercício 1

```python
df = pd.read_csv('diamonds.csv')
amostraCut = df.groupby('cut', group_keys=False).apply(lambda x: x.sample(frac=0.20))
amostraCut['cut'].value_counts()

```

### Exercício 2

```python
amostraProporcional = df.groupby('cut', group_keys=False).apply(lambda x: x.sample(frac=20000 / len(df)))
amostraProporcional['cut'].value_counts(normalize=True)
```

### Exercício 3

```python
amostraProporcional = df.groupby('cut', group_keys=False).apply(lambda x: x.sample(frac=20000 / len(df)))
df = pd.read_csv('diamonds.csv')

plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
sns.histplot(df['carat'], bins=30, kde=True)
plt.title("Distribuição de 'carat' no Dataset Original")

plt.subplot(3, 1, 2)
sns.histplot(amostraCut['carat'], bins=30, kde=True)
plt.title("Distribuição de 'carat' na Amostra Aleatória de 20% por 'cut'")

plt.subplot(3, 1, 3)
sns.histplot(amostraProporcional['carat'], bins=30, kde=True)
plt.title("Distribuição de 'carat' na Amostra Proporcional de 20.000 Entradas")

plt.tight_layout()
plt.show()
```

### Análise dos Resultados
Após gerar as amostras e plotar os histogramas da variável carat para o conjunto de dados original e para as duas amostras, observei que os resultados foram análogos. Aqui está uma análise detalhada de cada um:

##### Dataset Original: 
O histograma completo dos dados representa a distribuição total dos valores de carat na população. Ele serve como referência para verificar se as amostras refletem a distribuição geral.

##### Amostra Aleatória de 20% por cut: 
A amostra aleatória de 20% para cada categoria de cut mostrou uma distribuição semelhante à do dataset original. Isso indica que a amostragem aleatória simples conseguiu capturar bem as características da população, mantendo uma boa representatividade dos dados.

##### Amostra Proporcional de 20.000 Entradas:
A amostra proporcional também manteve uma distribuição similar ao conjunto de dados original, preservando a proporção de cada categoria na variável cut e a forma geral da distribuição dos valores de carat. Isso mostra que a amostragem proporcional foi eficaz para representar a população total.
