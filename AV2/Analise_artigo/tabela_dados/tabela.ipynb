```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "tabela.csv"  
data = pd.read_csv(file_path)

print(data.head())
print(data.info())

experiencia = data["Qual é o seu nível de experiência em programação?"].value_counts()
print("Distribuição por nível de experiência:")
print(experiencia)

frequencia_uso = data["Com que frequência você utiliza IA para aprender ou resolver problemas de programação?"].value_counts()
print("Frequência de uso de IA:")
print(frequencia_uso)

ferramentas = data["Se sim, qual(is) ferramenta(s) de IA você usou?"].value_counts()
print("Ferramentas de IA mais utilizadas:")
print(ferramentas)

habilidades = data["Quais habilidades de programação você considera que a IA ajuda a desenvolver? (Selecione todas que se aplicam)"].str.get_dummies(sep=", ").sum()
print("Habilidades impactadas:")
print(habilidades)

plt.figure(figsize=(10, 6))
sns.barplot(x=experiencia.index, y=experiencia.values, palette="viridis")
plt.title("Distribuição por Nível de Experiência")
plt.xlabel("Nível de Experiência")
plt.ylabel("Contagem")
plt.show()

plt.figure(figsize=(8, 8))
frequencia_uso.plot.pie(autopct="%.1f%%", startangle=90, cmap="viridis")
plt.title("Frequência de Uso de IA")
plt.ylabel("")
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=ferramentas.index, y=ferramentas.values, palette="viridis")
plt.title("Ferramentas de IA Mais Utilizadas")
plt.xlabel("Ferramentas")
plt.ylabel("Contagem")
plt.xticks(rotation=45, ha="right")
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=habilidades.index, y=habilidades.values, palette="viridis")
plt.title("Habilidades de Programação Impactadas pela IA")
plt.xlabel("Habilidades")
plt.ylabel("Contagem")
plt.xticks(rotation=45, ha="right")
plt.show()

output_path = "analise_estatisticas_aprendizado.csv"
habilidades.to_csv(output_path, index=True)
print(f"Relatório exportado para {output_path}")
```
