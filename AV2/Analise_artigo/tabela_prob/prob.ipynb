```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('analiseProbs.csv')

print("Tabela de Resumo:")
print(data)

categoria_count = data["Categoria"].value_counts()   
plt.figure(figsize=(8, 5))
categoria_count.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Distribuição de Erros por Categoria")
plt.xlabel("Categoria de Erro")
plt.ylabel("Frequência")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

linguagem_count = data["Linguagem"].value_counts()  
plt.figure(figsize=(8, 5))
linguagem_count.plot(kind="bar", color="lightgreen", edgecolor="black")
plt.title("Distribuição por Linguagem")
plt.xlabel("Linguagem")
plt.ylabel("Frequência")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
