```python
#exercício 1
from scipy.stats import binom

def moedaViciada(jogadas, probabilidade_coroa):
    esperado_coroas = jogadas * probabilidade_coroa
    return esperado_coroas

jogadas = 1000
probabilidade_coroa = 0.4 
resultado_problema_1 = moedaViciada(jogadas, probabilidade_coroa)
print(f"Resultado esperado de coroas: (algo em torno de {resultado_problema_1})")

#estimando números de caras e coroas
p = 0.6
n = 1000
cara = binom.rvs(n, p,size = 1)
coroa = n - cara
cara1 = n - coroa
print(f"número de: {coroa[0]} e o número de cara é {cara1[0]}")
```

```python
#exercício 2
from scipy.stats import binom
def probOvos(n, probabilidade_quebrar, max_quebras):
    probabilidade = binom.cdf(max_quebras, n, probabilidade_quebrar)
    return probabilidade

n = 20  
probabilidade_quebrar = 0.1  #exemplo de probabilidade 
max_quebras = 2 
resultado_problema_2 = probOvos(n, probabilidade_quebrar, max_quebras)
print(f"Probabilidade de2 ovos quebrados: {resultado_problema_2:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

#parametros gerais
n = 10  
p = 0.4  

#1- distribuição com todas as probabilidades de x = 0, 1, 2, ..., n
x_values = np.arange(0, n + 1) 
probabilidades = binom.pmf(x_values, n, p)  
print("Questão 1:\n")
print("Distribuição completa das probabilidades:")
for x, prob in zip(x_values, probabilidades):
    print(f"P(X = {x}) = {prob:.4f}")

#2- probabilidade de exatamente 2 pessoas terem sangue do tipo A
probabilidade_exatamente_2 = binom.pmf(2, n, p)
print("\nQuestão 2:\n")
print(f"\nProbabilidade de exatamente 2 pessoas terem sangue do tipo A: {probabilidade_exatamente_2:.4f}")

#3) gráfico de barras 
print("\nQuestão 3:\n")
plt.bar(x_values, probabilidades)
plt.xlabel('Número de pessoas com sangue tipo A')
plt.ylabel('Probabilidade')
plt.title('Distribuição Binomial (n=10, p=0.4)')
plt.show()

#4- probabilidade de no máximo 3 pessoas terem sangue do tipo A
probabilidade_no_maximo_3 = binom.cdf(3, n, p)
print("\nQuestão 4:\n")
print(f"\nProbabilidade de no máximo 3 pessoas terem sangue do tipo A: {probabilidade_no_maximo_3:.4f}")
```
