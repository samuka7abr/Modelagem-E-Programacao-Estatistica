```python
import pandas as pd
dados = pd.read_csv('TabelaSalarial.csv')
dados.head()
dados.describe()

#media, mediana, desvio, variancia
media = dados.mean()
mediana = dados.median()
devio = dados.std()
variancia = dados.var()
moda = dados.mode()

print(f"Médias: \n{media}\n")
print(f"Mediana: \n{mediana}\n")
print(f"Desvio padrão: \n{devio}\n")
print(f"Variância: \n{variancia}\n")
print(f"Moda: \n{moda}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
dados = pd.read_csv('TabelaSalarial.csv')
dados.head()
dados.describe()
dados.hist()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('TabelaSalarial.csv')

dados.boxplot()
plt.show()
```

#### Medidas de Centralidade:
- As médias mostram o valor médio baseado nas notas, a mediana mostra a centralidade dos dados (50% dos dados estão abaixo do valor da mediana.) 

#### Medidas de Dispersão:
- O desvio padrão e a variância mostram o quanto os dados variam em relação à média. A tabela apresenta um desvio padrão irregular contabilizando todas as variáveis, baixo em algumas alto em outras.

#### Distribuições (Histogramas):
- Os histogramas mostram a frequência de valores em intervalos específicos, ajudando a entender a distribuição dos dados. Revelam se os dados são simétricos ou não.

#### Boxplots:
- Os boxplots ajudam a visualizar a variação dos dados e a identificar outliers, que são valores que se destacam por serem muito diferentes dos demais. Um boxplot assimétrico pode indicar que os dados estão inclinados para um lado.

#### Conclusão:
- Essa análise oferece uma visão geral dos dados, utilizando estatísticas simples para entender a centralidade e a variação, visualizando as distribuições e identificando outliers. Isso é fundamental para tomar decisões informadas.
