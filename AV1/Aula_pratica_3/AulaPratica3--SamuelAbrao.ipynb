# Relação das variáveis: Salário X Idade:

### Primeiro bloco de código:
#### No primeiro bloco foi calculado a média dos valores de salário (em salários mínimos) nos intervalos de idade pré-definidos.


```python
import pandas as pd

dados = pd.read_csv('TabelaSalarial.csv')

dados['FaixaEtaria'] = pd.cut(dados['Idade (em anos)'], bins=[20, 30, 40, 50, 60, 70], labels=['20-30', '31-40', '41-50', '51-60', '61-70'])

tabela_salario_idade = dados.groupby('FaixaEtaria')['Salario (x Sal Min)'].mean()
print(tabela_salario_idade)
```

### Segundo Bloco de código:
#### já no segundo bloco, foi plotado um gráfico do tipo scatter plot para que a variável "Idade" fosse relacionada com o valor dos salários, mostrando com uma linha de tendência os valores de direção geral das variáveis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dados = pd.read_csv('TabelaSalarial.csv')

plt.figure(figsize=(10, 6))
sns.regplot(x='Idade (em anos)', y='Salario (x Sal Min)', data=dados, scatter_kws={'s': 50}, line_kws={'color': 'purple'})
plt.title('Relação entre Idade e Salário com Linha de Tendência')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.show()


```

# Relação das variáveis: Salário X N de Filhos:

### Primeiro bloco de código:
#### No primeiro bloco de código mostra um gráfico que apresenta a relação das médias salariais de acordo com o número de filhos (0 a 5)
#### Deixando claro o valor das médias


```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('TabelaSalarial.csv')

media_salario_por_filhos = dados.groupby('N de Filhos')['Salario (x Sal Min)'].mean()

plt.figure(figsize=(10, 6))

media_salario_por_filhos.plot(kind='bar', color='purple', edgecolor='black')

plt.ylim(0, 25)
plt.title('Média Salarial por Número de Filhos')
plt.xlabel('Número de Filhos')
plt.ylabel('Salário Médio (em Salários Mínimos)')
plt.show()
```

### Segundo bloco de código:
#### No segundo bloco de código foi plotado um gráfico do tipo boxplot para que os valores máximos e mínimos fossem considerados na análise.
#### Explicação do Boxplot: 
##### - O Boxplot tem um tamanho definido pela linha preta (sendo o tamanho de cima o valor máximo e o tamanho de baixo o valor mínimo naquela categoria [numero de filhos, nesse caso])

##### - A caixa azul é onde tem a maior concentração desses dados (em 0 filhos, está entre 8 e 13 salários)

##### - A linha verde dentro das caixas azuis é a média (ela não fica sempre exatamente no meio pq os dados podem ser distribuídos mais pra um lado doq pra o outro)

```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('TabelaSalarial.csv')

plt.figure(figsize=(10, 6))
dados.boxplot(column='Salario (x Sal Min)', by='N de Filhos', grid=True, showmeans=True)
plt.ylim(0, 25)
plt.title('Distribuição Salarial por Número de Filhos')
plt.suptitle('')  
plt.xlabel('Número de Filhos')
plt.ylabel('Salário (em Salários Mínimos)')
plt.show()
```

# Relação das variáveis: Salário X Grau de instituição


### No gráfico plotado foi mostrado a relação das médias salariais por grau de instituição. Nota-se que a média é maior para os funcionários que possuem ensino superior completo. É necessário que seja calculado o desvio padrão dessa relação para que a média não exclua valores extremos e mínimos.

```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('TabelaSalarial.csv')

salario_por_instrucao = dados.groupby('Grau de Instrução')['Salario (x Sal Min)'].mean()
salario_por_instrucao = salario_por_instrucao.sort_values()
desvio_padrao_por_instrucao = dados.groupby('Grau de Instrução')['Salario (x Sal Min)'].std()


plt.figure(figsize=(10, 6))
salario_por_instrucao.plot(kind='bar', color=['purple', 'violet', 'pink'])  
plt.ylim(0, 25)
plt.title('Evolução de Salários por Grau de Instrução')
plt.xlabel('Grau de Instrução')
plt.ylabel('Salário Médio (em Salários Mínimos)')
plt.show()

print(desvio_padrao_por_instrucao)


```

# Relação das variáveis: Salário X Região de Procedência:
### Primeiro Bloco de código:
#### No primeiro gráfico as informações mostradas fazem referencia à relação da média salarial pela região dos funcionários (Capital, Interior, Outra)
### Segundo Bloco de código:
#### No segundo bloco de código foi plotado um gráfico de boxplot para que todos os valores fossem considerados, incluindo a média, a tendẽncia e os valores extremos.

```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('TabelaSalarial.csv')

media_salario_por_regiao = dados.groupby('Região de Procedência')['Salario (x Sal Min)'].mean()


plt.figure(figsize=(10, 6))
media_salario_por_regiao.plot(kind='bar', color='skyblue', edgecolor='black')

plt.ylim(0, 25)
plt.title('Média Salarial por Região')
plt.xlabel('Região de Procedência')
plt.ylabel('Salário Médio (em Salários Mínimos)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('TabelaSalarial.csv')

plt.figure(figsize=(10, 6))

dados.boxplot(column='Salario (x Sal Min)', by='Região de Procedência', showmeans=True)

plt.ylim(0, 25)
plt.title('Distribuição Salarial por Região')
plt.suptitle('')  
plt.xlabel('Região de Procedência')
plt.ylabel('Salário (em Salários Mínimos)')
plt.show()
```
