### Histograma das variáveis

```python
# Estado Civil com matplotlib

estadocivil=[ dados['Estado Civil']]
plt.figure(figsize=(4, 3))
plt.ylabel('Frequência Absoluta')
plt.hist(estadocivil, bins=3)
# plt.grid(True)
plt.show()
```

#### Para as variáveis quantitativas

```python
# Com matplotlib

nfilhos=[dados['N de Filhos']]
plt.figure(figsize=(4, 3))
plt.xlabel('Quantidade de filhos')
plt.ylabel('Frequência Absoluta')
plt.hist(nfilhos, 5, rwidth=0.9)
plt.grid(True)
plt.show()

salario=[ dados['Salario (x Sal Min)']]
plt.figure(figsize=(4, 3))
plt.xlabel('Salário (x Salário Mínimo)')
plt.ylabel('Frequência Absoluta')
plt.hist(salario, 6, rwidth=0.9)
plt.grid(True)
plt.show()

idades=[ dados['Idade (em anos)']]
plt.figure(figsize=(4, 3))
plt.xlabel('Idade')
plt.ylabel('Frequência Absoluta')
plt.hist(idades, 6, rwidth=0.9)
plt.grid(True)
plt.show()
```
