### 1. o que faz o código acima (dados.describe())?

.describe() gera um resumo estatístico descritivo para as colunas do documento.csv.

###  2. Quais informações são apresentadas?
As informações apresentadas são:

* count: O número de entradas não nulas para cada coluna.
* mean: A média dos valores em cada coluna.
* std: O desvio padrão, que mede a dispersão dos valores em torno da média.
* min: O valor mínimo encontrado na coluna.
* 25%: O primeiro quartil, que é o ponto abaixo do qual 25% dos dados caem.
* 50% (mediana): O valor central dos dados (50% dos dados estão abaixo deste valor).
* 75%: O terceiro quartil, que é o ponto abaixo do qual 75% dos dados caem.
* max: O valor máximo encontrado na coluna.

### 3. Todas as variáveis têm o mesmo número de informações?
Não, nem todas as variáveis tem as mesmas informações.
N. de filhos tem menos informaçõoes que as outras variáveis.

### 4. O que significa a variável "mean" e "std"?
Mean: média
std: desvio padrão (ou dispersão) dos valores em torno da média


### 5.  Considerando o o máximo, o mínimo, a média e o "std", o que podemos inferir sobre as variáveis nº de filhos, salário e idade?
* N. de filhos: 
    - std: Apresenta um desvio padrão baixo (em comparação com os outros) (1,268028)
    - mean: Apresenta uma média de 1.65 filhos 
    - min e max: Apresenta um mínimo de 0 filhos e uma máxima de 5 filhos.
* Salário:
    - std: Apresenta um desvio padrão médio (em comparação com os outros) (4.587458	)
    - mean: Apresenta média de 11,12(dizima) salários
    - min e max: Apresenta um mínimo de 4 salários e um máximo de 23.3
* Idade:
    - std: Apresenta um alto desvio padrão (6.737422)
    - mean: Apresenta média de 34.583333
    -min e max:  Apresenta um mínimo de 20 anos e um máximo de 48 anos


### 6. Por que não aparecem as variáveis "Grau de instrução" e "Região de Procedência"?
Não aparecem porque .describe() só fornece estatísticas para colunas numéricas.

### 7. Considerando todas as análises e informações coletadas, a média é um bom indicativo para os dados?
A média é um bom indicativo, mas deve ser considerada em conjunto com outras estatísticas, como a mediana e o desvio padrão, especialmente porque as variáveis, como salário, podem ter grandes desvios que influenciam a média
